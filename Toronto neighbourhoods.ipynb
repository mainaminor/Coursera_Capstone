```python
#import BeautifulSoup library and MAKE SURE TO IMPORT REQUESTS
from bs4 import BeautifulSoup as bs
import requests
```

```python
#define a variable "source" which is the text of the html code for the wikipedia page. 
#Define "soup" as the text formatted to lxml.
#use the "prettify" function to reproduce the indents in the html file (and make it easier to read :)

source=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=bs(source,'lxml')
print(soup.prettify())
```

I inspect the elements on the actual web page to work out which HTML element contains the body of the table

I discover it is the "tbody" element

So I specify a new variable "tab" which will find the first (and only!) instance of this element.

```python
tab = soup.find('tbody').text
print(tab)
```

I discover this is a string separated by tabs. Oh my

I import pandas and numpy, and a new one called "re" and get straight to work

```python
import pandas as pd
import numpy as np
```

```python
tablist = tab.split('\t') #I am splitting the string into an list based on tab (symbolized by "'\t'")
tablist
```

My list has only one (very long) element. But i notice the delimiter "\n" is being used. So i do one more split to get a multi-element list

```python
tablist2=tablist[0].split('\n') 
len(tablist2)
print(tablist2)
```

Now we are cooking! Let me first check the length of my new list

```python
len(tablist2)
```

I notice that there is a space before "Postcode" and one after "neighbourhood", so in fact instead of 3 columns, i need to re-shape my data to 5 columns.

But 1446 is not divisible by 5?

So think, why not just see what happens if i define an array based on 1445 elements

```python
x=np.array(tablist2[0:1445])
```

I do a bit of math and work out if i have 5 columns, i must have 289 rows (to get to 1445 elements). So i define a new array based on re-shaping x

```python
y=x.reshape(289,5)
y.shape
```

```python
y
```

Now I define a dataframe based on this array

```python
df=pd.DataFrame(data=y[1:,0:], columns=y[0,0:])
```

Next step: Work out which rows have Borough being not assigned

```python
indexNames = df[ df['Borough'] =='Not assigned'].index
```

```python
indexNames
```

Drop those rows

```python
df.drop(indexNames , inplace=True)
```

For some reason I had to re-set the index because the new table did not automatically do that. so for example my first row had an index of 2?? Very weird

```python
df=df.reset_index(drop=True)
df
```

This next part was insane, and it took some Googling. 

Essentially there is an apply function that lets you group the column values. Don't ask me how it works

I do 3 things:

<ul>
  <li>Define a new dataframe which is <b> just </b> the postcodes with the neighbourhoods joined</li>
  <li>Define another dataframe which is the unique mapping of postcodes to boroughs</li>
  <li>Join these based on postcode</li>
</ul>


```python
#Step 1

df2=df.groupby('Postcode')['Neighbourhood'].apply(lambda x: "%s" % ', '.join(x))
df2
```

```python
#Step 2

df3=df[['Postcode','Borough']].drop_duplicates()
```

```python
#Step 3
result = pd.merge(df3,df2, on='Postcode')
result
```

I am going to be a stickler here and rename the postcode to Postal Code

```python
result.rename(columns={"Postcode":"Postal Code"}, inplace=True)
```

Notice now there is a neighbourhood which is "not assigned".

We need to replace this with the "borough" name

How do we do this?

Approximately 2 hours of Google later, here is what i found.

There is a numpy function called <b> numpy.where </b>
    
Watch this

```python
result['Neighbourhood'] = np.where(result['Neighbourhood'] == "Not assigned", result['Borough'], result['Neighbourhood'])
```

```python
result
```

<b> BOOM </b> 

Job done

Now lets add the co-ordinates

```python
coord=pd.read_csv('https://cocl.us/Geospatial_data')  
```

```python
coord.head()
```

```python
result_f = pd.merge(result,coord, on='Postal Code', how="left")
result_f.head()
```

Define a new dataframe to be used as the basis for clustering the results by lat and longitude

```python
result_clustering = result_f.drop(['Postal Code','Borough','Neighbourhood'], 1)
result_clustering
```

Run k means to cluster by location. Display the results

```python
#import kMeans
from sklearn.cluster import KMeans

kclusters=5

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(result_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Insert Cluster latels

```python
result_f.insert(0, 'Cluster Labels', kmeans.labels_)
result_f
```

Import folium and matplotlib to generate map

```python
import folium # map rendering library

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# create map
map_clusters = folium.Map(location=[43.7, -79.3832], zoom_start=11)
```

```python
# set color scheme for the clusters
xs = np.arange(kclusters)
ys = [i + xs + (i*xs)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
#define and generate map

for lat, lon, poi, cluster in zip(result_f['Latitude'], result_f['Longitude'], result_f['Neighbourhood'], result_f['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)

map_clusters
```
